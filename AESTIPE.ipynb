```python
def flip_bit(block, bit_index):
    block = np.array(block, dtype=np.uint8)
    byte_i = bit_index // 8
    bit_i = bit_index % 8
    flat = block.flatten()
    flat[byte_i] ^= (1 << bit_i)
    return flat.reshape(4, 4)

def avalanche_score(r, n=1000):
    scores = []
    for _ in range(n):
        P = np.random.randint(0, 256, (4, 4), dtype=np.uint8)
        bit = np.random.randint(0, 128)
        P_prime = flip_bit(P, bit)
        C = OneBlockAES(P, r)
        C_prime = OneBlockAES(P_prime, r)
        C_bytes = np.array(C, dtype=np.uint8).flatten()
        C_prime_bytes = np.array(C_prime, dtype=np.uint8).flatten()
        diff = sum(bin(a ^ b).count('1') 
                   for a, b in zip(C_bytes, C_prime_bytes))
        scores.append(diff / 128 * 100)
    return np.mean(scores)

import matplotlib.pyplot as plt

rounds = range(1, 11)
scores = [avalanche_score(r) for r in rounds]

plt.plot(rounds, scores, marker='o', color='blue')
plt.axhline(y=50, color='red',label='Seuil demandé (50%)')
plt.xlabel('Nombre de tours')
plt.ylabel('Effet Avalanche (%)')
plt.title('Effet Avalanche en fonction du nombre de tours — AES-128')
plt.xticks(range(1, 11))
plt.legend()
plt.show()
```
